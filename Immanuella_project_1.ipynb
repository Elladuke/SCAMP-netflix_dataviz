### Question 2
Guess The Number: Write a program where the computer randomly generates a number between 0 and 20. The user needs to guess what the number is. If the user guesses wrong, tell them their guess is either too high, or too low. 

```python
import random
```

```python
computer_num = random.randint(0,20)
```

```python
while True:
    guess = (input('Make a guess: '))
    try:
        guess = float(guess)
        if (guess < computer_num):
            print('Too low!')
        elif (guess > computer_num):
            print('Too high!')
        else:
            print('You got it right!')
            break
    
    except ValueError:
        print("Guess a number instead.")
        continue
   
```

### Question 3
Write a program, which generates a random password for the user. The password should be a minimum of 6 characters long.
Ask the user how long they want their password to be, and how many letters and numbers they want in their password. Have a mix of upper and lowercase letters, as well as numbers and symbols. 

```python
import string
```

```python
def generate_pwd(pwd_length, letters_length, numbers_length):
    numbers = random.sample(range(0,9), numbers_length)
    alphabets = string.ascii_letters
    symbols = '!@#$%^&*<>?'
    letters_chars = random.sample(alphabets+symbols,letters_length)
    password =  letters_chars + [str(n) for n in numbers]
    
    return "".join(password)  
```

```python
password = ""
pwd_length =  eval(input('How long do you want your password to be (min of 6 chars): '))
letters_length =  eval(input('How many letters/symbols?: '))
numbers_length =  eval(input('How many numbers?: '))

while pwd_length < 6 or letters_length+numbers_length != pwd_length:
    print('Make sure the length is greater than 6 and the numbers and letters length adds to the password length')
    if pwd_length < 6:
        pwd_length =  eval(input('How long do you want your password to be (min of 6 chars): '))
    if letters_length+numbers_length != pwd_length:
        letters_length =  eval(input('How many letters/symbols?: '))
        numbers_length =  eval(input('How many numbers?: '))

password = generate_pwd(pwd_length, letters_length, numbers_length)
print()
print('Your password is ', password)
```
